```python
#Script for reading the data
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

data=[]
for i in [20,48,75,95,99]:
    url = 'Team_1/data/Ref_Factor_'+str(i)+'.csv'
    data.append(pd.read_csv(url))
```

```python
#important labels
wvl_label = 'Wvl'
Rad_label = 'Rad. (Ref.)'
```

```python
#creates a list of data frames with just Wvl and Rad
shortdata = []
for i in range(5):
    shortdata.append(pd.DataFrame({'Wvl':data[i][wvl_label],'Rad':data[i][Rad_label]}))
```

```python
plt.plot(shortdata[0].Wvl,shortdata[0].Rad,'b-')
plt.plot(shortdata[1].Wvl,shortdata[1].Rad,'g-')
plt.plot(shortdata[2].Wvl,shortdata[2].Rad,'r-')
plt.plot(shortdata[3].Wvl,shortdata[3].Rad,'k-')
plt.plot(shortdata[4].Wvl,shortdata[4].Rad,'y-')
```
