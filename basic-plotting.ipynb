```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = []

for i in [20,48,75,95,99]:
    path = 'data/Ref_Factor_' + str(i) +'.csv'
    data.append(pd.read_csv(path))
    
#Thanks to Santiago for the for loop
```

```python
plt.figure()
plt.plot(data[0]['Wvl'], data[0]['Rad. (Ref.)'], 'r.')
plt.plot(data[1]['Wvl'], data[1]['Rad. (Ref.)'], color='black', marker ='.')
plt.plot(data[2]['Wvl'], data[2]['Rad. (Ref.)'], 'y.')
plt.plot(data[3]['Wvl'], data[3]['Rad. (Ref.)'], 'g.')
plt.plot(data[4]['Wvl'], data[4]['Rad. (Ref.)'], 'b.')
plt.show
```

```python
n = 11 
plt.figure()
plt.plot(data[0]['Wvl'], pd.rolling_mean(data[0]['Rad. (Ref.)'], n), 'r-')
plt.plot(data[1]['Wvl'], pd.rolling_mean(data[1]['Rad. (Ref.)'], n), 'k-')
plt.plot(data[2]['Wvl'], pd.rolling_mean(data[2]['Rad. (Ref.)'], n), 'y-')
plt.plot(data[3]['Wvl'], pd.rolling_mean(data[3]['Rad. (Ref.)'], n), 'g-')
plt.plot(data[4]['Wvl'], pd.rolling_mean(data[4]['Rad. (Ref.)'], n), 'b-')
plt.show()
```
